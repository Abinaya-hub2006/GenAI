```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer, AutoConfig
from scipy.special import softmax

# Load model and tokenizer
model_path = "cardiffnlp/twitter-roberta-base-sentiment-latest"
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForSequenceClassification.from_pretrained(model_path)
```

```python
def sentiment_analysis(text):
    encoded_input = tokenizer(text, return_tensors='pt')
    output = model(**encoded_input)
    scores_ = output.logits.detach().numpy()[0]
    scores_ = softmax(scores_)
    labels = ['Negative', 'Neutral', 'Positive']
    scores = {label: float(score) for label, score in zip(labels, scores_)}
    return scores
```

```python
import gradio as gr

# Create Gradio interface
demo = gr.Interface(
    fn=sentiment_analysis,
    inputs=gr.Textbox(placeholder="Write your text here..."),
    outputs="label",
    examples=[
        ["I'm thrilled about the job offer!"],
        ["The weather today is absolutely beautiful."],
        ["I'm so frustrated with this software glitch."],
        ["The customer service was terrible at the store."]
    ],
    title="Sentiment Analysis App",
    description="This app classifies text as Positive, Neutral, or Negative."
)

# Launch the app
demo.launch()
```

**PARTS OF SPEECH**

```python
import gradio as gr
from transformers import pipeline

# Load POS tagging model from Hugging Face
pos_pipeline = pipeline("token-classification", model="vblagoje/bert-english-uncased-finetuned-pos")

def process_text(text):
    results = pos_pipeline(text)
    formatted = [(ent["word"], ent["entity"]) for ent in results]
    return {"text": text, "entities": formatted}

# Create Gradio interface
demo = gr.Interface(
    fn=process_text,
    inputs=gr.Textbox(label="Input Text", placeholder="Enter sentence here..."),
    outputs=gr.HighlightedText(label="POS Tags",
                              show_legend=True),
    examples=[
        ["The quick brown fox jumps over the lazy dog"],
        ["I love learning about natural language processing"],
        ["Gradio makes machine learning demo creation simple!"]
    ],
    title="Real-Time Part-of-Speech Tagger",
    description="Enter text to see instant POS tagging results using BERT",
    allow_flagging="never"
)

demo.launch()

```

```python
from transformers import pipeline

# Load the POS tagging pipeline
try:
    pos_pipeline = pipeline("token-classification", model="vblagoje/bert-english-uncased-finetuned-pos")
except Exception as e:
    print(f"Error loading model: {str(e)}")
    exit()

def process_text(text):
    if not text.strip():
        return "Error: Input text cannot be empty."
    try:
        results = pos_pipeline(text)
        formatted_output = []
        for result in results:
            word = result["word"]
            entity = result["entity"]
            formatted_output.append(f"{word}/{entity}")
        return " ".join(formatted_output)
    except Exception as e:
        return f"Error during processing: {str(e)}"

# Main loop for command-line input
if __name__ == "__main__":
    while True:
        text = input("Enter text (or type 'exit' to quit): ")  # Get input from the user
        if text.lower() == "exit":
            break
        tagged_text = process_text(text)
        print(tagged_text)
```

**CHATBOT**


```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load the model and tokenizer
try:
    model_name = "microsoft/DialoGPT-medium"
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForCausalLM.from_pretrained(model_name)
    print("Chatbot model loaded successfully!")
except Exception as e:
    print(f"Error loading model: {str(e)}")
    exit()

# Function to generate a response
def respond(user_input, chat_history_ids=None):
    try:
        # Encode user input and append it to the chat history
        new_input_ids = tokenizer.encode(user_input + tokenizer.eos_token, return_tensors="pt")
        bot_input_ids = (
            torch.cat([chat_history_ids, new_input_ids], dim=-1) if chat_history_ids is not None else new_input_ids
        )

        # Generate a response
        chat_history_ids = model.generate(
            bot_input_ids,
            max_length=1000,
            pad_token_id=tokenizer.eos_token_id,
            temperature=0.7,
            top_k=50,
            top_p=0.95,
        )

        # Decode the response and return it along with updated chat history
        response = tokenizer.decode(chat_history_ids[:, bot_input_ids.shape[-1]:][0], skip_special_tokens=True)
        return response, chat_history_ids
    except Exception as e:
        return f"Error generating response: {str(e)}", chat_history_ids

# Main loop for command-line interaction
if __name__ == "__main__":
    import torch

    print("Hello! I am your chatbot. Type 'exit' to end the conversation.")
    chat_history_ids = None  # Initialize conversation history

    while True:
        user_input = input("You: ")
        if user_input.lower() == "exit":
            print("Chatbot: Goodbye!")
            break

        # Get chatbot response
        response, chat_history_ids = respond(user_input, chat_history_ids)
        print(f"Chatbot: {response}")
```

**PARAPHRASER**

```python
from transformers import pipeline

# Load the paraphrasing pipeline
try:
    paraphraser = pipeline("text2text-generation", model="Vamsi/T5_Paraphrase_Paws")
    print("Paraphrasing model loaded successfully!")
except Exception as e:
    print(f"Error loading model: {str(e)}")
    exit()

def paraphrase(text):
    try:
        # Generate paraphrase
        paraphrased_text = paraphraser(text, max_length=100, num_return_sequences=1)[0]['generated_text']
        return paraphrased_text
    except Exception as e:
        return f"Error generating paraphrase: {str(e)}"

# Main loop for command-line interaction
if __name__ == "__main__":
    print("Welcome to the Text Paraphraser! Type 'exit' to end the session.")

    while True:
        user_input = input("Enter text to paraphrase: ")
        if user_input.lower() == "exit":
            print("Goodbye!")
            break

        # Get paraphrased text
        result = paraphrase(user_input)
        print("Paraphrased Text:", result)
```

**DOCUMENT ANALISER**

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def collect_documents():
    print("Enter your documents (type 'exit' to finish):")
    documents = []

    while True:
        user_input = input("Document: ")
        if user_input.lower() == "exit":
            break
        documents.append(user_input)

    return documents

def retrieve_documents(query, vectorizer, tfidf_matrix, df):
    # Transform the query into TF-IDF format using the same vectorizer
    query_tfidf = vectorizer.transform([query])

    # Compute cosine similarity between the query and documents
    cosine_similarities = cosine_similarity(query_tfidf, tfidf_matrix).flatten()

    # Get indices of documents sorted by similarity score
    related_docs_indices = cosine_similarities.argsort()[::-1]

    # Return sorted documents with their similarity scores
    return [(df.iloc[i]["Document"], cosine_similarities[i]) for i in related_docs_indices]

if __name__ == "__main__":
    # Collect documents from the user
    user_documents = collect_documents()

    if user_documents:
        # Create a DataFrame to hold the documents
        df = pd.DataFrame(user_documents, columns=["Document"])

        # Initialize the TF-IDF Vectorizer and fit on collected documents
        vectorizer = TfidfVectorizer()
        tfidf_matrix = vectorizer.fit_transform(df["Document"])

        print("Documents collected successfully!")

        while True:
            user_query = input("Enter your query (type 'exit' to quit): ")
            if user_query.lower() == "exit":
                print("Goodbye!")
                break

            # Retrieve and display relevant documents
            results = retrieve_documents(user_query, vectorizer, tfidf_matrix, df)
            print("\nRelevant Documents:")
            for doc, score in results:
                print(f"Document: {doc} | Similarity Score: {score:.4f}")
    else:
        print("No documents were entered.")
```

**TEXT GENERATOR**

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

# Load pre-trained model and tokenizer
model_name = "gpt2"  # You can also use "gpt2-medium", "gpt2-large", or "gpt2-xl"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

# Function to generate text
def generate_text(prompt, max_length=100):
    # Encode the input prompt
    input_ids = tokenizer.encode(prompt, return_tensors='pt')

    # Generate text using the model
    with torch.no_grad():
        output = model.generate(input_ids, max_length=max_length, num_return_sequences=1,
                                no_repeat_ngram_size=2,
                                early_stopping=True)

    # Decode the generated text
    generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
    return generated_text

if __name__ == "__main__":
    print("Welcome to the Text Generator! Type 'exit' to quit.")

    while True:
        user_input = input("Enter your prompt: ")
        if user_input.lower() == "exit":
            print("Goodbye!")
            break

        # Generate and display text
        generated_output = generate_text(user_input)
        print("\nGenerated Text:")
        print(generated_output)
```
