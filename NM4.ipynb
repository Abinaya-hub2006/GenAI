```python
from transformers import AutoTokenizer, AutoModelForQuestionAnswering

# Load BioBERT model and tokenizer
model_name = "dmis-lab/biobert-large-cased-v1.1-squad"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForQuestionAnswering.from_pretrained(model_name)

# Function to answer questions
def answer_question(context: str, question: str) -> str:
    inputs = tokenizer.encode_plus(question, context, return_tensors="pt")
    outputs = model(**inputs)
    answer_start = outputs.start_logits.argmax()
    answer_end = outputs.end_logits.argmax() + 1
    answer = tokenizer.convert_tokens_to_string(
        tokenizer.convert_ids_to_tokens(inputs.input_ids[0][answer_start:answer_end])
    )
    return answer

# Get user input
context = input("Enter medical context or article: ")
question = input("Enter your medical question: ")

# Get and display the answer
answer = answer_question(context, question)
print("\n💡 Answer:", answer)
```

**WORKING ON ALL VIDEO AUDIO AND TEXT**

```python
# Install required libraries
!pip install -q sentence-transformers transformers faiss-cpu librosa opencv-python

# Imports
import numpy as np
import faiss
import librosa
import cv2
import torch
from PIL import Image
from transformers import (
    AutoTokenizer,
    AutoModel,
    AutoModelForSeq2SeqLM,
    Wav2Vec2Processor,
    Wav2Vec2Model,
    CLIPProcessor,
    CLIPModel,
    pipeline
)
from sentence_transformers import SentenceTransformer

# Device setup
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

# ===================== Embedding Models ====================== #

# Text
text_model = SentenceTransformer('all-MiniLM-L6-v2')
text_model.to(device)

# Image / Video (first frame) with CLIP
clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32").to(device)
clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")

# Audio (Wav2Vec2)
audio_processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-base-960h")
audio_model = Wav2Vec2Model.from_pretrained("facebook/wav2vec2-base-960h").to(device)

# Small LLM for generation (local)
gen_tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-base")
gen_model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-base").to(device)

# ===================== Embedding Functions ====================== #

def embed_text(text):
    vec = text_model.encode(text)
    return vec.astype("float32")

def embed_image(path):
    image = Image.open(path).convert("RGB")
    inputs = clip_processor(images=image, return_tensors="pt").to(device)
    with torch.no_grad():
        vec = clip_model.get_image_features(**inputs).squeeze().cpu().numpy()
    return vec.astype("float32")

def embed_audio(path):
    y, sr = librosa.load(path, sr=16000)  # Wav2Vec2 expects 16kHz
    inputs = audio_processor(y, sampling_rate=16000, return_tensors="pt", padding=True).to(device)
    with torch.no_grad():
        vec = audio_model(**inputs).last_hidden_state.mean(dim=1).squeeze().cpu().numpy()
    return vec.astype("float32")

def embed_video(path):
    cap = cv2.VideoCapture(path)
    ret, frame = cap.read()
    cap.release()
    if ret:
        img = Image.fromarray(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
        inputs = clip_processor(images=img, return_tensors="pt").to(device)
        with torch.no_grad():
            vec = clip_model.get_image_features(**inputs).squeeze().cpu().numpy()
        return vec.astype("float32")
    else:
        return np.zeros(512, dtype="float32")

# Normalize/resize vectors to match FAISS dimension
def normalize_vector(vec, target_dim):
    vec = vec.reshape(-1)
    if vec.shape[0] != target_dim:
        vec = np.resize(vec, target_dim)
    return vec.astype('float32')

# ===================== FAISS Vector Store ====================== #

class VectorStore:
    def __init__(self, dim):
        self.index = faiss.IndexFlatL2(dim)
        self.data = []

    def add(self, vec, meta):
        vec = np.asarray(vec).reshape(1, -1).astype("float32")
        self.index.add(vec)
        self.data.append(meta)

    def search(self, vec, top_k=3):
        vec = np.asarray(vec).reshape(1, -1).astype("float32")
        D, I = self.index.search(vec, top_k)
        return [self.data[i] for i in I[0] if i < len(self.data)]

# ===================== Generation Function ====================== #

def generate_response(prompt, context):
    full_prompt = f"{prompt}\nContext:\n" + "\n".join(context)
    inputs = gen_tokenizer(full_prompt, return_tensors="pt", truncation=True).to(device)
    with torch.no_grad():
        outputs = gen_model.generate(**inputs, max_new_tokens=100)
    return gen_tokenizer.decode(outputs[0], skip_special_tokens=True)

# ===================== Setup and Usage ====================== #

# Initialize vector store with target dim
target_dim = 512
store = VectorStore(dim=target_dim)

# Add items
store.add(normalize_vector(embed_text("A man riding a horse in the desert."), target_dim), "Text: Horse riding description")
store.add(normalize_vector(embed_image("/content/R.jfif"), target_dim), "Image: Desert horse image")
store.add(normalize_vector(embed_audio("/content/istockphoto-1903666781-640_adpp_is.mp4"), target_dim), "Audio: Horse neighing")
store.add(normalize_vector(embed_video("/content/mixkit-scared-horse-neighing-85.wav"), target_dim), "Video: Horse galloping")

# ===================== RAG Function ====================== #

def query_multimodal(query):
    q_vec = normalize_vector(embed_text(query), target_dim)
    retrieved = store.search(q_vec, top_k=3)
    print("Retrieved Contexts:")
    for item in retrieved:
        print("-", item)
    print("\nGenerated Answer:")
    print(generate_response(query, retrieved))

# ===================== Run a Sample Query ====================== #

query_multimodal("What is the man doing?")
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

# Load tokenizer and model
model_name = "EleutherAI/pythia-410m"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Use GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
```

```python
def chat_with_bot(prompt, history=None, max_length=150):
    if history is None:
        history = ""

    input_text = history + f"User: {prompt}\nBot:"
    inputs = tokenizer.encode(input_text, return_tensors="pt").to(device)

    outputs = model.generate(
        inputs,
        max_length=inputs.shape[1] + max_length,
        pad_token_id=tokenizer.eos_token_id,
        do_sample=True,
        top_k=50,
        top_p=0.95,
        temperature=0.7,
    )

    output_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    # Get only the bot's latest reply
    reply = output_text[len(input_text):].strip().split("User:")[0].strip()

    # Return updated history
    return reply, input_text + reply + "\n"

# Start Chat
history = ""
while True:
    user_input = input("You: ")
    if user_input.lower() in ["exit", "quit"]:
        break
    response, history = chat_with_bot(user_input, history)
    print("Bot:", response)
```

**YOUTUBE VIDEO SUMMARIZER**

```python
from youtube_transcript_api import YouTubeTranscriptApi
from transformers import pipeline

# Extract video ID from URL
def extract_video_id(url):
    if "watch?v=" in url:
        return url.split("watch?v=")[-1].split("&")[0]
    elif "youtu.be/" in url:
        return url.split("youtu.be/")[-1].split("?")[0]
    return url

# Step 1: Fetch transcript
video_url = "https://www.youtube.com/watch?v=dQw4w9WgXcQ"  # Replace with your link
video_id = extract_video_id(video_url)
transcript_list = YouTubeTranscriptApi.get_transcript(video_id)
transcript = " ".join([t['text'] for t in transcript_list])

# Step 2: Summarize using Transformers
summarizer = pipeline("summarization", model="google/pegasus-xsum")

summary = summarizer(transcript, max_length=100, min_length=30, do_sample=False)[0]['summary_text']
print("📌 Summary:\n", summary)
```
