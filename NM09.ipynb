```python
# Install necessary libraries
!pip install diffusers transformers torch scipy moviepy

from diffusers import StableDiffusionPipeline
import torch
from moviepy.editor import ImageSequenceClip
import os

# Initialize the Stable Diffusion model (Ensure CUDA is available for GPU support)
model = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16)
model.to("cuda")  # Use GPU if available

# Function to generate images from text prompts
def generate_images_from_text(prompts, output_folder="generated_images"):
    # Create output directory if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Generate images for each prompt and save them
    image_paths = []
    for i, prompt in enumerate(prompts):
        # Generate image
        image = model(prompt).images[0]

        # Save image to disk
        image_path = os.path.join(output_folder, f"image_{i+1}.png")
        image.save(image_path)
        image_paths.append(image_path)
        print(f"Image {i+1} saved at {image_path}")

    return image_paths

# Text prompts to generate images from
prompts = [
    "A futuristic cityscape with flying cars",
    "A beautiful beach at sunset",
    "A mountain range with clouds",
    "A dense forest with sunlight filtering through the trees"
]

# Generate images
image_paths = generate_images_from_text(prompts)

# Create a video from the generated images
def create_video_from_images(image_paths, output_video="output_video.mp4", fps=1):
    # Create a video clip from the images
    clip = ImageSequenceClip(image_paths, fps=fps)

    # Write the video to a file
    clip.write_videofile(output_video, codec="libx264")
    print(f"Video saved at {output_video}")

# Generate the video
create_video_from_images(image_paths)

# Optionally, you can download the video from Colab if you're working there
from google.colab import files
files.download("output_video.mp4")
```

**STORY GENERATOR**


```python
# Install Hugging Face's Transformers library
!pip install transformers torch

# Import necessary modules from Hugging Face
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load pre-trained GPT-2 model and tokenizer from Hugging Face
model_name = "gpt2"  # You can use "gpt2-medium" or "gpt2-large" for a larger model

tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

# Move model to GPU if available
import torch
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Function to generate a story from a prompt
def generate_story(prompt, max_length=300, temperature=0.7):
    # Encode the prompt text into tokens
    inputs = tokenizer.encode(prompt, return_tensors="pt").to(device)

    # Generate text
    outputs = model.generate(
        inputs,
        max_length=max_length,
        temperature=temperature,  # Controls the creativity: higher is more creative
        num_return_sequences=1,  # You can change this to generate multiple stories
        no_repeat_ngram_size=2,  # Avoid repeating the same n-grams
        top_k=50,  # Controls the diversity of generated words
        top_p=0.95,  # Controls the cumulative probability for top-p sampling
        do_sample=True,  # Sample from the output distribution
        pad_token_id=tokenizer.eos_token_id
    )

    # Decode the output tokens to text
    story = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return story

# Example prompt for story generation
prompt = "Once upon a time in a faraway kingdom, there lived a young princess named Rose."

# Generate a story
story = generate_story(prompt)
print(story)
```
