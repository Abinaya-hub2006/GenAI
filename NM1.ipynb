**LANGUAGE TRANSLATOR**

```python
from transformers import MarianMTModel, MarianTokenizer

# Function to load the model and perform translation
def translate_text(text, src_language, tgt_language):
    # Construct model name dynamically based on user input
    model_name = f'Helsinki-NLP/opus-mt-{src_language}-{tgt_language}'

    # Load the model and tokenizer
    model = MarianMTModel.from_pretrained(model_name)
    tokenizer = MarianTokenizer.from_pretrained(model_name)

    # Tokenize input text and perform translation
    translated = model.generate(**tokenizer(text, return_tensors="pt", padding=True, truncation=True))
    translation = tokenizer.decode(translated[0], skip_special_tokens=True)

    return translation

# Function to get language code based on user-friendly input
def get_language_code(language):
    # Define a mapping of language names to model codes
    language_codes = {
        'english': 'en',
        'spanish': 'es',
        'french': 'fr',
        'german': 'de',
        'italian': 'it',
        'portuguese': 'pt',
        'chinese': 'zh-cn',
        'japanese': 'ja',
        'russian': 'ru',
        'arabic': 'ar',
        'hindi': 'hi',
    }

    # Return the language code for the selected language
    return language_codes.get(language.lower(), 'en')

# List of supported languages for the user to choose from
supported_languages = [
    "english", "spanish", "french", "german", "italian",
    "portuguese", "chinese", "japanese", "russian",
    "arabic", "hindi"
]

# Display available languages
print("Supported Languages:")
for index, lang in enumerate(supported_languages, 1):
    print(f"{index}. {lang.capitalize()}")

# User input for source and target languages
src_language_input = input("Enter the source language (e.g., English): ").lower()
dest_language_input = input("Enter the target language (e.g., Spanish): ").lower()

# Check if entered languages are valid
if src_language_input not in supported_languages or dest_language_input not in supported_languages:
    print("Invalid language input. Please choose from the supported languages.")
else:
    # Convert to language codes
    src_language_code = get_language_code(src_language_input)
    dest_language_code = get_language_code(dest_language_input)

    # User input for text to translate
    text_to_translate = input("Enter the text you want to translate: ")

    # Perform the translation
    translated_text = translate_text(text_to_translate, src_language_code, dest_language_code)

    # Output the result
    print(f"\nTranslated text: {translated_text}")
```
